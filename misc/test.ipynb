```python
import csv

asta_file_path = 'Astaadhyaayii.txt'
out_csv = 'out.csv'

with open(asta_file_path, 'r', encoding='utf-8') as f:
    asta_data = f.readlines()
```

```python
out_map_csv = 'out_map.csv'

def char_label_mapping(text, labels):
    text_chars = [char_t for char_t in text]
    marked_list = [char_i for char_i in labels]
    result = []
    
    for i, char in enumerate(marked_list):
        if char == "+":
            result.append(' SP')
        else:
            result.append(' NSP')
    
    return ''.join(result)

with open(out_map_csv, 'w', newline='', encoding='utf-8') as csvfile:
    csv_writer = csv.writer(csvfile)
    
    csv_writer.writerow(['Text', 'Marked', 'Labels'])  
    
    for line in asta_data:
        if '\t=>\t' in line:
            text, marked = line.split('\t=>\t')
            print(text.strip())
            print(marked.strip())
            labels = char_label_mapping(text.strip(), marked.strip())
            csv_writer.writerow([text.strip(), marked.strip(), labels])
        else:
            print(f"Skipping line due to incorrect format: {line}")
```

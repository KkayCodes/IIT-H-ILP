```python
malay_file_path = 'malayalam_dataset'  
output_csv = 'output.csv' 

with open(malay_file_path, 'r', encoding='utf-8') as f:
    malay_data = f.readlines()

def char_label_mapping(text, labels):
    text_chars = text.split()  
    label_list = labels.split() 
    
    result = []
    temp = []
    
    for i, char in enumerate(text_chars):
        temp.append(char)
        if label_list[i] == "SP":
            result.append(''.join(temp) + "+")
            temp = []
        else:
            result.append(''.join(temp))
            temp = []
    
    if temp:
        result.append(''.join(temp))
    
    return ''.join(result)

with open(output_csv, 'w', newline='', encoding='utf-8') as csvfile:
    csv_writer = csv.writer(csvfile)
    
    csv_writer.writerow(['Text', 'Labels', 'Mapped_Text'])
    
    for line in malay_data:
        text, labels = line.split('\t')
        mapped_text = char_label_mapping(text.strip(), labels.strip())
        csv_writer.writerow([text.strip(), labels.strip(), mapped_text])
```

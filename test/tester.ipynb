```python
# Import required libraries
import pytesseract

# Configure Tesseract to use the Sanskrit model ('san')
pytesseract.pytesseract.tesseract_cmd = '/usr/bin/tesseract'  # Adjust path if necessary

# Set language to Sanskrit model (assuming the model is installed)
LANGUAGE = 'san'

# Function to perform OCR on an image
def ocr_image(image_path):
    # Perform OCR using Tesseract with the Sanskrit model
    text = pytesseract.image_to_string(image_path, lang=LANGUAGE)
    return text

# Example usage
# Replace 'path/to/your/image.png' with the path to your input image
image_path = '/home/kkay/IIT-H/OCR/ILP-OCR/test/complex/c4.png'
extracted_text = ocr_image(image_path)

# Display the OCR result
print("Extracted Text:")
print(extracted_text)
```
